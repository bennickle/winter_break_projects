As a slightly more complicated example, let's use the Pascal's Triangle program from earlier to see how it fits in the Jupyter environment:

```python
# Simple Pascal's triangle test code to practice GitHub / Jupyter basics

def factorial(n):
    num = 1
    for i in range(n):
        num *= (i+1)
    return num

def choose(n, k):
    return int(factorial(n)/(factorial(k)*factorial(n-k)))

def binomial(N):
    for i in range(N+1):
        numlist = []
        for j in range(i+1):
            numlist.append(choose(i,j))
        print(numlist)

rownum = int(input("How many rows? "))
binomial(rownum)
```

# Git and GitHub Introduction:

Git is a free and open source version control system. A version control system allows you to save your code as you update it and track your progress, which is very useful for collaboration and debugging. 

GitHub is an website where you can post your code in a directory called a repository. Other people can then access your code.

General GitHub workflow:
- Edit Files / Write Code
- Stage Files
- Commit (Local Save)
- Push (Save to GitHub)


Let's now walk through common Git commands:

```python
# Clone lets you access the code locally:

git clone git@github.com:bennickle/winter_break_projects.git

<#
If instead you had a normal folder that you wanted to turn into a 
git reporsitory, you would use the following command:
#>

git init
# This will turn whatever working directory you're in into a git repo
```

Once you either make changes to a cloned repo or initialize a directory into a git repo, you need to let git know which files to track. To check the current progress/checking of the repository, use the following:

```python
git status
```

git status will tell you whether you have files in the repository that are modified but not saved, as well as untracked files that have been created since the last commit.

In addition, git status will tell you which branch you are on, and how that branch compares to the main branch.

After using git status, you can make sure that the git is tracking what you want with the following commands:

```python
# All files in repository:
git add .

# Specific files:
git add filename_here
```

The next step to saving your progress is to commit your changes. When committing, you have to leave a title and a description so other people (or your future self) know what changes you made. The general format for committing is the following:

```python
git commit -m "Title" -m "Message"
```

This saves the changes to your git repo locally, but to update the repository for everyone you must push the repo to GitHub with the following command:

```python
git push origin main
```
