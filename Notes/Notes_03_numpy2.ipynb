Next, let's look at doing derivatives and integrals:

For derivatives, np.gradient(y,x) gives you the an array of values where the ith element gives you a numerical approximation for the derivative at the the x value x[i].

For integrals, we can use the cumulative sum function, cumsum, which will take in an array and return a new array of the same size where each entry is a sum of all of the elements up to the point you're at. If you scale this by x[1] = x[0], this gives a Riemann sum, so you can approximate integrals quickly with this.


```python
x = np.linspace(1,10,100)
y = 1/x**2 + np.sin(x)
dydx = np.gradient(y,x) # gives derivative
y_int = np.cumsum(y) * (x[1]-x[0]) # gives integral where y(0) = 0

plt.plot(x,y, label="y = f(x)") # Plots function
plt.plot(x,dydx, label="y = f'(x)") # Plots derivative
plt.plot(x,y_int, label="y = int(f(x))") # Plots integral
plt.legend(loc="upper right") # Adds legend and chooses its location
```

Indexing in more dimensions is trickier. First index gives rows, second index gives columns, just like a matrix in linear algebra.

When we're doing multivariable functions, we can't just use linspace for x and y then do something like x^2 + y^2, since that will add the arrays element wise. Instead, we need to make a meshgrid to get everything:

```python
x = np.linspace(0,10,1000)
y = np.linspace(0,10,1000)

xv, yv = np.meshgrid(x,y)

xv
```

```python
zv = xv**2 + yv**2

plt.contourf(xv, yv, zv, levels=30)
plt.colorbar()
```

```python
# For a 3d plot:

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.plot_surface(xv, yv, zv, cmap='viridis')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

Question 1: 

Let f(x,y) = e^(-x^2 + y^2) * sin(x) for x in [-2,2] and y in [-2,2]. 

1. Make a contour plot of f
2. Find the volume |f(x,y)| in the specificed x and y range
3. Find the volume |f(x,y)| in only in the region within 0.5 of the origin

```python
# Problem 1
x = np.linspace(-2,2,1000)
y = np.linspace(-2,2,1000)
xv, yv = np.meshgrid(x, y)

fv = np.exp(-(xv**2 + yv**2)) * np.sin(xv)

plt.contourf(xv,yv,fv, levels=100)
plt.colorbar()
```

```python
# Problem 2: volume

"""
Since we don't care about signed volume, we need to take the absolute value first.
Next, since fv is a 2d array, summing it will give a 1d array. We need two sums
(a double integral) to get one value. We then multiply by dy dx.

"""


sum(sum(np.abs(fv))) * (x[1]-x[0]) * (y[1]-y[0])
```

```python
"""
Alternatively, we can use the ravel method:

"""

sum(np.abs(fv.ravel())) * np.diff(x)[0] * np.diff(y)[0]
```

In the above solution, using ravel makes it so the sum only needs to be applied once. Next np.diff(x) creates an array of the first difference of x. Because we used linspace, this will be a constant valued array. Accessing any element of np.diff(x) will give us our differential.

```python
# Problem 3:

fv[(xv**2 + yv**2) >= 0.5**2] # This gives f values outside the disk

sum(np.abs(fv[(xv**2 + yv**2) >= 0.5**2].ravel()))* np.diff(x)[0] * np.diff(y)[0]
```
